```python
import numpy as np
import pandas as pd
import tensorflow as tf
import glob
import matplotlib.pyplot as plt
```

```python
train_files_N=glob.glob("D:/chest_xray/train/NORMAL/*")
train_files_Y=glob.glob("D:/chest_xray/train/PNEUMONIA/*")
```

```python
train_files=[]
for i, trf in enumerate(train_files_N):
    tk=[trf,0]
    train_files.append(tk)
```

```python
import re
def index(s,ch):
    a=[]
    for m in re.finditer(ch, s):
         a.append(m.start()) 

    return a
```

```python
for i,trf in enumerate(train_files_Y):
    idx=index(trf,'_')
    if(trf[idx[1]+1]=='b'):
        tk=[trf,1]
    else:
        tk=[trf,2]
    train_files.append(tk)
    
```

```python
train_files
```

```python
import random
random.shuffle(train_files)
```

```python
train_files
```

```python
import cv2
X_train=np.zeros((len(train_files),180,180,3))
Y_train=np.zeros((len(train_files),3))
for i, trf in enumerate(train_files):
    ims = cv2.imread(trf[0])
    #ims1=cv2.cvtColor(ims, cv2.COLOR_BGR2GRAY)
    #print(ims.shape)
    #break
    ress = cv2.resize(ims, dsize=(180, 180), interpolation=cv2.INTER_CUBIC)
 
    #print(ress.shape)
    X_train[i,:,:,:]=ress
    Y_train[i, trf[1]] = 1
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
X_train[0]
```

```python
X_train=X_train/255.0
```

```python
X_train[0]
```

```python
X_train=X_train.reshape(len(X_train),180,180,3)
```

```python
from keras.models import Sequential
```

```python
model=Sequential()

model_part1=tf.keras.applications.ResNet50(
    include_top=False,
    weights="imagenet",
    input_tensor=None,
    input_shape=(180,180,3),
    pooling='avg',
    classes=3,
)
for layer in model_part1.layers:
    layer.trainable=False
```

```python
from keras.layers import Flatten,Dense
```

```python
model.add(model_part1)
model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='Adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=10, validation_split=0.1, initial_epoch=0)
```
